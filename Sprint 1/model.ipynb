```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
```

```python
more_than_once = pd.read_csv("../datasets/trusted/more_than_once.csv", index_col=0)
more_than_once.head()
```

```python
sold_products = pd.read_csv("../datasets/trusted/sold_products.csv", index_col=0)
sold_products.head()
```

```python
customers_history = pd.merge(
    more_than_once, sold_products, on='order_id'
).groupby(
    ['customer_unique_id', 'product_category_name']
).agg('count')['order_id'].reset_index()

customers_history.head()
```

```python
customers_category_count = customers_history.groupby(
    ['customer_unique_id']
).agg(['count'])['product_category_name']['count'].reset_index()
customers_category_count.head()
```

```python
more_than_one_cat = customers_category_count[customers_category_count['count'] > 1]['customer_unique_id']
more_than_one_cat.head()
```

```python
filtered_customers = customers_history[customers_history['customer_unique_id'].isin(more_than_one_cat)]
filtered_customers.head()
```

```python
onehot = pd.DataFrame(
    columns=sold_products['product_category_name'].sort_values().unique(),
    index=filtered_customers['customer_unique_id'].unique()
).fillna(0)

for customer_id, category in filtered_customers[['customer_unique_id', 'product_category_name']].values:
    onehot.loc[customer_id, category] = 1

onehot.head()
```
