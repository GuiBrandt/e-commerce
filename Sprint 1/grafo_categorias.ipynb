```python
import pandas as pd
import json

from IPython.core.display import display, HTML
from string import Template
```

```python
category_sales_table = pd.read_csv("./category_sales_table.csv", index_col=0)
category_sales_table.head()
```

## Grafo de relaÃ§Ãµes entre categorias

```python
%%javascript

require.config({
    paths: {
        "d3": "https://d3js.org/d3.v5.min"
    }
});

require(["d3"], d3 => window.d3 = d3);
```

```python
with open("./graph/template.js", "r") as file:
    js_template = Template(file.read())

with open("./graph/template.html", "r") as file:
    html_template = Template(file.read())
```

```python
scores = category_sales_table.sum() / category_sales_table.sum().sum()

nodes = []
for category in category_sales_table.index:
    nodes.append({
        'id': category,
        'weight': int(100 * scores[category] / scores.max())
    })
```

```python
links = []

for category in category_sales_table.index:
    rel = category_sales_table.loc[category]

    for key, value in dict(rel[rel != 0]).items():
        links.append({
            "source": category,
            "target": key,
            "value": int(value)
        })
```

```python
nodes_json = json.dumps(nodes)
links_json = json.dumps(links)

js_text = js_template.substitute({
    "categories": nodes_json,
    "category_links": links_json,
    "graphdiv": "graph-div"
})

html_text = html_template.substitute({
    "js_text": js_text
})

HTML(html_text)
```
